```python
import pandas as pd
from sklearn.model_selection import KFold
from sklearn.neighbors import KNeighborsClassifier
from sklearn.impute import SimpleImputer
from sklearn.metrics import accuracy_score

# Assuming 'is_genuine' is the target variable, and other features are predictors
X = df.drop('is_genuine', axis=1)
y = df['is_genuine']

# Handle missing values using SimpleImputer
imputer = SimpleImputer(strategy='mean')
X_imputed = imputer.fit_transform(X)

# Number of folds for cross-validation
num_folds = 5
kf = KFold(n_splits=num_folds, shuffle=True, random_state=42)

# Range of odd values for the number of neighbors
k_neighbors_range = range(3, 61, 2)

# List to store average accuracies for each value of k_neighbors
avg_accuracies = []

for k_neighbors in k_neighbors_range:
    print(f"Number of Neighbors: {k_neighbors}")

    # Initialize KNN classifier
    knn_classifier = KNeighborsClassifier(n_neighbors=k_neighbors)

    # List to store accuracies for each fold
    fold_accuracies = []

    # Perform k-fold cross-validation
    for train_index, test_index in kf.split(X_imputed):
        X_train, X_test = X_imputed[train_index], X_imputed[test_index]
        y_train, y_test = y.iloc[train_index], y.iloc[test_index]

        # Train the KNN model
        knn_classifier.fit(X_train, y_train)

        # Make predictions on the test set
        y_pred = knn_classifier.predict(X_test)

        # Evaluate the accuracy
        accuracy = accuracy_score(y_test, y_pred)
        fold_accuracies.append(accuracy)
        print(f'Accuracy for fold: {accuracy}')

    # Calculate and print the average accuracy for the current k_neighbors
    avg_accuracy = sum(fold_accuracies) / len(fold_accuracies)
    avg_accuracies.append(avg_accuracy)
    print(f'Average Accuracy for k_neighbors={k_neighbors}: {avg_accuracy}')
    print('-' * 50)  # Separating different values of k_neighbors
```

```python

# Plotting
plt.figure(figsize=(8, 6))
plt.plot( avg_accuracies, marker='o', linestyle='-', color='b')
plt.title('Average Accuracies for Experiments')
plt.xlabel('Experiments')
plt.ylabel('Average Accuracy')
plt.grid(True)
plt.show()
```

```python

import pandas as pd
from sklearn.model_selection import KFold
from sklearn.linear_model import LogisticRegression
from sklearn.impute import SimpleImputer
from sklearn.metrics import accuracy_score

# Assuming 'is_genuine' is the target variable, and other features are predictors
X = df.drop('is_genuine', axis=1)
y = df['is_genuine']

# Handle missing values using SimpleImputer
imputer = SimpleImputer(strategy='mean')
X_imputed = imputer.fit_transform(X)

# Number of folds for cross-validation
num_folds = 5
kf = KFold(n_splits=num_folds, shuffle=True, random_state=42)

# List to store average accuracies for each fold
avg_accuracies = []

# Range of regularization parameters (C values) for logistic regression
C_values = [0.001, 0.01, 0.1, 1, 10, ]

for C_value in C_values:
    # print(f"Regularization Parameter (C): {C_value}")

    # Initialize logistic regression classifier
    logistic_reg_classifier = LogisticRegression(C=C_value, random_state=42)

    # List to store accuracies for each fold
    fold_accuracies = []

    # Perform k-fold cross-validation
    for train_index, test_index in kf.split(X_imputed):
        X_train, X_test = X_imputed[train_index], X_imputed[test_index]
        y_train, y_test = y.iloc[train_index], y.iloc[test_index]

        # Train the logistic regression model
        logistic_reg_classifier.fit(X_train, y_train)

        # Make predictions on the test set
        y_pred = logistic_reg_classifier.predict(X_test)

        # Evaluate the accuracy
        accuracy = accuracy_score(y_test, y_pred)
        fold_accuracies.append(accuracy)
        # print(f'Accuracy for fold: {accuracy}')

    # Calculate and print the average accuracy for the current C_value
    avg_accuracy = sum(fold_accuracies) / len(fold_accuracies)
    avg_accuracies.append(avg_accuracy)
    # print(f'Average Accuracy for C={C_value}: {avg_accuracy}')
    # print('-' * 50)  # Separating different values of C_value
```

```python
# Plotting
plt.figure(figsize=(8, 6))
plt.plot( avg_accuracies, marker='o', linestyle='-', color='b')
plt.title('Average Accuracies for Experiments')
plt.xlabel('Experiments')
plt.ylabel('Average Accuracy')
plt.grid(True)
plt.show()
```
